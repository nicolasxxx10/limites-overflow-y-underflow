```python
import math


print("===== FLOATS (precision doble) ====")

#Overflow
x = 1.0
while not math.isinf(x):
    last = x
    x *= 2

maxima_potencia_dos = last

x = maxima_potencia_dos
incremento = maxima_potencia_dos / 2

while incremento > 0:
    if not math.isinf(x + incremento):
        x = x + incremento
    incremento /= 2

print("Limite antes del Overflow:", x)


#Underflow
y = 1.0
while y!= 0.0:
      last_small = y
      y /= 2

print("Limite antes del Underflow:", last_small)
```

```python
import math


print("======== ENTEROS (64 BITS) ========")

BITS = 64
MIN = -2**(BITS-1)
MAX = 2**(BITS-1)-1

def sum(a, b):
    resultado = a + b

    if resultado > MAX:
        print("Limite antes del Overflow:", MAX)
    elif resultado < MIN:
        print("Limite antes del Underflow:", MIN)
    return resultado


# Probando un caso que causa overflow
prueba_overflow = sum(MAX, 1)

# Probando un caso que causa underflow
prueba_underflow = sum(MIN, -1)
```
