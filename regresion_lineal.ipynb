```python
#Importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt


# Datos del sistema solar desde Mercurio hasta Neptuno aqui hacemos de los datos una lista para poder usarlos como variables


# Distancia media al Sol (AU).
a = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.19, 30.07])


# Periodo orbital (años)
T = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8])


# Transformación logarítmica


x = np.log(a)
y = np.log(T)


# Ajuste lineal (mínimos cuadrados) utilizando la funcion ya creada en numpy


coef = np.polyfit(x, y, 1)


B = coef[0]      # pendiente
logA = coef[1]   # intercepto


A = np.exp(logA)


print("Resultados del ajuste")
print("Pendiente B =", B)
print("Constante A =", A)


# Curva ajustada
x_fit = np.linspace(min(x), max(x), 100)
y_fit = B * x_fit + logA


# Gráfica

plt.scatter(x, y, label="Datos planetarios")
plt.plot(x_fit, y_fit, label="Ajuste lineal")


plt.xlabel("log(a)")
plt.ylabel("log(T)")
plt.title("Verificación de la tercera ley de Kepler")
plt.legend()


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Considerando la siguiente lista de datos 
data = {'x': [0, 0.020, 0.036, 0.056, 0.073, 0.091, 0.109, 0.128, 0.148, 0.167, 0.184, 0.205, 0.221],
        'F': [0, 0.199, 0.394, 0.588, 0.786, 0.984, 1.158, 1.355, 1.552, 1.748, 1.944, 2.138, 2.336]}
df = pd.DataFrame(data)      

plt.figure(figsize=(10,5))
plt.scatter(df['x'], df['F'])
plt.xlabel('Elongation')
plt.ylabel('Force')
plt.grid(True)
plt.show()

display (df)
```

```python

#Elongacion promedio
sum_x = 0
i = 0
n = len(df['x'])

while i <= n-1:
  sum_x = sum_x + df['x'][i]
  i = i+1

prom_x = sum_x/n
print(f"Elongacion promedio: {prom_x:.3f}")

#Fuerza promedio 
sum_y = 0
i = 0
n = len(df['F'])

while i <=n-1:
  sum_y = sum_y+df ['F'][i]  
  i = i+1

prom_y = sum_y / n
print(f"Fuerza promedio: {prom_y:.3f}")

#Covarianza entre x y F
cov = 0
i = 0
n = len(df)

while i <=n-1:
  cov = cov+(df['x'][i]-prom_x)*(df['F'][i]-prom_y)
  i = i+1
print(f"Covarianza entre x y F: {cov:.3f}")

#Varianza de x
var = 0
i = 0
n = len(df)

while i<=n-1:
    var = var+(df['x'][i]-prom_x)**2
    i = i+1
print(f"Varianza de x: {var:.3f}")

```

```python
#Obtener los parametros de la recta y el coeficiente de determinacion

m = cov/var
b = prom_y-(m*prom_x)

#calcular suma de errores cuadrados
suma_error2 = 0
suma_total2 = 0

i = 0
while i<=n-1:
    
    y_pred = m*df['x'][i] + b   # valor predicho
    error = df['F'][i] - y_pred
    
    suma_error2 += error**2
    suma_total2 += (df['F'][i] - prom_y)**2
    
    i = i+1

R2 = 1 - (suma_error2 / suma_total2)

#Mostrar resultados
print(f"El parametro m es {m:.3f}")
print(f"El parametro b es {b:.3f}")
print(f"El coeficiente de determinacion es {R2:.4f}")

print('========================')

print(f"La ecuacion de la recta mediante regresion lineal es: F={m:.3f}x+{b:.3f}")
```
