```python
import matplotlib.pyplot as plt

x, y = [], []

with open("datos_lcg.txt") as f:
    for line in f:
        a, b = map(int, line.split())
        x.append(a)
        y.append(b)

plt.scatter(x, y, s=10)
plt.xlabel("r_i")
plt.ylabel("r_{i+1}")
plt.title("Residuos potenciales")
plt.savefig('residuos_potenciales.png')
plt.show()
```
