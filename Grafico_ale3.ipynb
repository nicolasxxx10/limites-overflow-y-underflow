```python
import matplotlib.pyplot as plt

x, y = [], []

with open("datos_rand.txt") as f:
    for line in f:
        a, b = map(int, line.split())
        x.append(a)
        y.append(b)

plt.scatter(x, y, s=10)
plt.xlabel("x_i")
plt.ylabel("x_{i+1}")
plt.title("Mersenne Twister")
plt.savefig('mersenne_twister.png')
plt.show()
```
