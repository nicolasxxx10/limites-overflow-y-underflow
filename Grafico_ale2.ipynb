```python
import matplotlib.pyplot as plt

x = []
y = []

with open("datos.txt") as f:
    for line in f:
        a, b = map(int, line.split())
        x.append(a)
        y.append(b)

plt.scatter(x, y)
plt.xlabel("x_i")
plt.ylabel("r_{i+1}")
plt.title("Pares sucesivos")
plt.savefig('Pares')
plt.show()
```
